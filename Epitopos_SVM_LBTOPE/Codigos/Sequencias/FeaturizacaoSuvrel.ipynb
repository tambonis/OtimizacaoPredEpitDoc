```python
#Pacotes

import warnings
warnings.filterwarnings("ignore", message="numpy.dtype size changed")
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")

import pickle
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Função Suvrel

def Suvrel(tab, classe, gamma):

    tab = np.column_stack([tab, classe])

    K = 2.0

    N = tab.shape[1] - 1
    epslon = np.zeros((N,N))

    for u in range(N):
        for v in range(N):
            mu = np.mean(tab[(tab[:,-1] == 1)][:,u]), np.mean(tab[(tab[:,-1] == -1)][:,u])
            mv = np.mean(tab[(tab[:,-1] == 1)][:,v]), np.mean(tab[(tab[:,-1] == -1)][:,v])

            epslon[u,v] = (2.0 - (K-1.0)*gamma)*(np.cov(tab[(tab[:,-1] == 1)][:,u], tab[(tab[:,-1] == 1)][:,v])[0][1] + \
                      np.cov(tab[(tab[:,-1] == -1)][:,u], tab[(tab[:,-1] == -1)][:,v])[0][1]) - \
                      gamma*K*K*np.cov(mu,mv)[0][1]


    epslon = (-1.0*epslon)/np.sqrt(np.sum(epslon*epslon))

    #Teste de verificação. 
    #print("Teste de verificacao (Somatorio do quadrado dos pesos): ", np.sum(epslon*epslon))

    return(epslon)
```

```python
## Função para cálculo da raiz do single value decomposition aplicado no tensor.

def svd_epslon(X, y, kgamma):
    
    epslon = Suvrel(tab=X, classe=y, gamma=kgamma)
    u, s, vh = np.linalg.svd(epslon, full_matrices=True)
    s = np.sqrt(s)
    epslon = np.dot(u[:, :(epslon.shape[1])] * s, vh)
    #epslon[epslon<0]=0
    
    return(s, epslon)
```

```python
def dists(data, ydata):
    
    rplus = data[ydata==1,:]
    rless = data[ydata==-1,:]

    rsumplus = np.zeros(rplus.shape[0])
    for j in range(rplus.shape[0]):
        for i in range(rplus.shape[1]):
            rsumplus[j] = rsumplus[j] + rplus[j,i]**2        
    rsumplus = np.sqrt(rsumplus)

    rsumless = np.zeros(rless.shape[0])
    for j in range(rless.shape[0]):
        for i in range(rless.shape[1]):
            rsumless[j] = rsumless[j] + rless[j,i]**2        
    rsumless = np.sqrt(rsumless)

    rsumplus_mean = np.mean(rsumplus)
    rsumless_mean = np.mean(rsumless)

    rplus_sigma = (rsumplus - rsumplus_mean)**2
    rplus_sigma = np.sqrt(np.sum(rplus_sigma))/(len(rplus_sigma))

    rless_sigma = (rsumless - rsumless_mean)**2
    rless_sigma = np.sqrt(np.sum(rless_sigma))/(len(rless_sigma))

    rplus_mean = data[ydata==1, :]
    rplus_mean = np.mean(rplus_mean, axis=0)

    rless_mean = data[ydata==-1, :]
    rless_mean = np.mean(rless_mean, axis=0)

    deltaX = np.sqrt(np.sum((rplus_mean - rless_mean)**2))
    deltaX = deltaX/(np.sqrt(rplus_sigma*rless_sigma))
    
    return(deltaX)
```

```python
def getbest(dftnp, ydft, kgamma):
    
    scaler = StandardScaler()
    
    dftnp_s = np.copy(dftnp)
    
    dftnp = scaler.fit_transform(dftnp)
    
    x0 = dists(data=dftnp, ydata=ydft)
    
    print("Gamma: ")
    print(kgamma)
    print("\n Distância sem Suvrel: ")
    print(x0)
    #dftnp = StandardScaler().fit_transform(dftnp)

    s, epslon = svd_epslon(X=dftnp, y=ydft, kgamma=kgamma)

    #Treino efetivo
    for i in range(dftnp_s.shape[0]):
        dftnp_s[i,:] = np.dot(epslon, dftnp_s[i,:])
    
    dftnp_s = scaler.fit_transform(dftnp_s)
    
    x1=dists(data=dftnp_s, ydata=ydft)
    
    print("\n Distância com Suvrel: ")
    print(x1)
    print("\n")
    print("#####################################")
    
    return(x1)
```

```python
#Carregar dados

with open('../Sequencias/DadosTreinoFeaturizados', 'rb') as fp:
        DadosTreino = pickle.load(fp)
    
y = np.array(DadosTreino['class'])
```

```python
## Parâmetros. 

rand_state = None #Semente fixa

selection = range(10, 150, 20)
#selection.append(127)
selection = np.array(selection)
selection = np.append(selection, [140, 150, 160, 170, 180, 200, 300, 400, 500])
selection 
```

```python
#Carregar dados 

kgamas_mais_altos = []
distancias_kgammas = []

gammas = np.linspace(start=1.5, stop=2.5, num=150)
for j in range(0, int(selection.shape[0])):
    
    X = DadosTreino.drop(['class'], 1)
    X = np.array(DadosTreino[DadosTreino.columns.values[0:selection[j]]])

    results = []
    for i in range(len(gammas)):
        
        print("Selecao: ", selection[j])
        print("Shape de X: ", X.shape)
        print("\n")
        
        results = np.append(results, getbest(dftnp=X, ydft=y, kgamma=gammas[i]))

    results_rmeans = np.array(pd.DataFrame(results).rolling(5, 
                                         center=True, min_periods=1).mean())
    
    kgamma = gammas[np.argmax(results_rmeans)]

    kgamas_mais_altos.append(kgamma)
    distancias_kgammas.append(results_rmeans[np.argmax(results_rmeans)])
```
